```python
import uproot
import numpy as np
import pandas as pd
from scipy import constants
import subprocess
import os
import matplotlib.pyplot as plt
import scipy.io as io
from scipy.interpolate import interp1d

starts = 1 #[um]
ends = 10
qs = 4
snum = 40
srange = np.exp(np.linspace(np.log(starts),np.log(ends),snum))
size = np.round(srange*2,decimals=2)
Emin = 120 #140 #800
Emax = 6.2e4 #5.9e4 #4750
Enum = 300 #10
Enum_add = 29
xi = np.arange(-Enum_add, Enum + 1, 1)
f = interp1d([1, Enum],
             [np.log(Emin), np.log(Emax)],
             kind='linear',
             fill_value='extrapolate',
             assume_sorted=True)
energyin = np.exp(f(xi))
sizein = size
```

```python
subprocess.call("rm -rf log.txt",shell=True)
subprocess.call("rm -rf test.mac",shell=True)
subprocess.call("rm -rf Record.root",shell=True)
```

```python
for Ei in range(0,300):
    for rhoi in range(len(sizein)):
        E = energyin[Ei]
        rho = sizein[rhoi]
        with open("test.txt","w") as f:
            f.write("/run/numberOfThreads 16\n")
            f.write("/B2/det/setTargetThick "+"{:.2f}".format(rho)+" um\n")
            f.write("/run/initialize\n")
            f.write("/control/verbose 0\n/run/verbose 0\n/event/verbose 0\n/tracking/verbose 0\n")
            f.write("/gun/energy "+"{:.4f}".format(E)+" keV\n")
            f.write("/run/beamOn 1000000")
        subprocess.call("mv test.txt test.mac",shell=True)
        subprocess.call("./exampleB2a test.mac >log.txt",shell=True)
        # --------modify the following path to your own path---------
        # subprocess.call("mv Record.root /Users/zfyin/Work/Code/Geant4Project/Ering/build/Record_"+"{:.4f}".format(E)+"_"+"{:.2f}".format(rho)+".root",shell=True)
        # subprocess.call("rm -rf log.txt",shell=True)
        # subprocess.call("rm -rf test.mac",shell=True)
```
